```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file = pd.read_csv('/content/drive/MyDrive/energydata_complete.csv')
file
```

```python
sub_file = file[['T2','T6']]
sub_file
```

```python
result = file[['Appliances']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(sub_file,result,test_size = 0.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
predicted_value = model.predict(x_test)
predicted_value
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, predicted_value)
score
```
